```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Optional, for a nicer plot
import numpy as np
from scipy import stats


df = pd.read_csv("C:\\Users\\Wael7\\Data Analytics Course\\Lab Work\\CleanDF.csv")
df.head(10)
```

```python
purchaseSumDf = df.copy()
purchaseSumDf['total_cost'] = purchaseSumDf['freight_value'] + purchaseSumDf['price']
purchaseSumDf.head(5)
```

the previous table shows the sum of the freight value and the price of the order, Using it to display the normal distribution of how much the average customer spends

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=purchaseSumDf[['total_cost','price','freight_value']])
plt.title('Distribution of purchases')
plt.xlabel('Order value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

The data looks extremely squeezed due to the outlier that is displaying anomalies or large purchases that make the data harder to read.</br> Setting an upper bound improve the data visualization.



```python
upperBound = purchaseSumDf['total_cost'].quantile(0.95)

plt.figure(figsize=(10, 6))
sns.kdeplot(data=purchaseSumDf[['total_cost','freight_value','price']])
plt.title('Distribution of purchases reduced')
plt.xlabel('Order value')
plt.ylabel('Density')
plt.grid(True)
plt.xlim(0,upperBound)
plt.show()
```

```python
upperBound = purchaseSumDf['total_cost'].quantile(0.95)

# Create the KDE plot
plt.figure(figsize=(10, 6))
sns.kdeplot(data=purchaseSumDf['total_cost'],color='red')
sns.kdeplot(data=purchaseSumDf['freight_value'],color='orange')
sns.kdeplot(data=purchaseSumDf['price'],color='green')

plt.title('Distribution of purchases ')
plt.xlabel('Order value')
plt.ylabel('Density')
plt.grid(True)

# Set the x-axis limits to focus on the data up to the upper bound
plt.xlim(0, upperBound)

# Add a vertical line at the median (50% probability)
median_value = purchaseSumDf[['total_cost','freight_value','price']].median()
plt.axvline(median_value[0], color='red', linestyle='--', label='average cart purchase')
plt.axvline(median_value[1],color='orange',linestyle='--', label='average freight fees')
plt.axvline(median_value[2],color='green',linestyle='--', label='average price of cart excluding freight')

# Show the plot
plt.legend()
plt.show()
```

From this data we can deduce that the average of a cart purchase is 100</br>
The data looks close to a normal distribution however we can normalize it to later compare it
The Box-Cox transformation is a family of power transformations that are used to stabilize the variance and make a dataset conform more closely to normality. It's particularly useful when you're working with statistical techniques that assume a normally distributed dataset.

```python

# Create a DataFrame with purchase data (example)


# Apply the Box-Cox transformation to make the data more like a normal distribution
purchaseSumDf['purchase_amount_transformed'], _ = stats.boxcox(purchaseSumDf['total_cost'])

# set upper bound, for this instance i did not add any 

# Plot the transformed distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=purchaseSumDf, x='purchase_amount_transformed', kde=True, bins=30)
plt.title('Normalized Cart purchase')
plt.xlabel('Transformed Purchase Amount')
plt.ylabel('Frequency')

median_value = purchaseSumDf['purchase_amount_transformed'].median()
upper25=purchaseSumDf['purchase_amount_transformed'].quantile(0.25)
lower25=purchaseSumDf['purchase_amount_transformed'].quantile(0.75)
plt.axvline(median_value, color='yellow', linestyle='--')
plt.axvline(upper25, color='green', linestyle='--')
plt.axvline(lower25, color='red', linestyle='--')
plt.grid(True)

# Set the x-axis limits to ignore the top 1%

plt.show()
```

This was the average cart purchase which is 100 including freight value. the lower 25% is represented by the lines behind the green lines. the upper 25% is shown in the area after the red line.

after that knowing the average of a cart purchase we would like to determine what are the most categories bought in the top 5 brazilian states in terms of orders.

```python

import matplotlib.pyplot as plt
df = pd.read_csv("C:\\Users\\Wael7\\Data Analytics Course\\Lab Work\\productDf.csv")
state_counts = df['customer_state'].value_counts()

# Select the top 5 states
top_states = state_counts.head(7).index

# Filter the DataFrame to only include the top 5 states
df = df[df['customer_state'].isin(top_states)]
df.head(10)
```

```python
state_counts = df['customer_state'].value_counts()
# Create a pie chart for the state occurrences
plt.figure(figsize=(8, 8))
plt.pie(state_counts, labels=state_counts.index, autopct='%1.1f%%')
plt.title('Pie chart of states with the most purchases')

# Display the pie chart
plt.show()
state_counts.plot(kind='bar', color='skyblue')
plt.ylabel('Sales')
plt.title('Bar chart of states with most purchases')
plt.show()
```

Sao paulo is the largest selling state in olist, accounting for almost 50% of sales of the whole website when compared to other states.</br> The population of Sao paulo is a factor contributing to this drastic difference when compared to other states, according to <a href='https://www.worldatlas.com/articles/brazilian-states-by-population.html'>worldatlas.com</a> the population of sao paulo is 44,035,304 in 2017, Which is 10% of the brazilian population.</br>
Now that we know the states that purchase the most, and the average cost of an order, It would better help us understand our customer by knowing what each states orders most


```python
grouped = df.groupby(['customer_state', 'product_category_name_english']).size().unstack().fillna(0)
for customer_state in grouped.index:
    categories = grouped.loc[customer_state]
    top_categories = categories.sort_values(ascending=False).head(10)
    # Create a pie chart for the top 20 categories in this state
    plt.figure(figsize=(6, 6))
    plt.pie(top_categories, labels=top_categories.index, autopct='%1.1f%%', startangle=140)
    plt.title(f'Top 10 Product Categories in {customer_state}')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
```
